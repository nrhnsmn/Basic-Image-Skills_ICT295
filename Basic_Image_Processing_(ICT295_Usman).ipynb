```python
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
from google.colab import drive

# Load an image from a file
image = cv2.imread('/content/drive/MyDrive/2023/ICT295 Computer Vision/Activity 2: Basic Image Skills/ID.jpg')

if image is None:
    print("Error: Could not open or find the image.")
else:
    # Display the image in a window
    cv2_imshow(image)
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Display the gray image in a window
cv2_imshow(gray_image)
```

```python
# Save the grayscale image as a file
cv2.imwrite('grayscale_image.jpg', gray_image)
```

```python
# Load the grayscale image
gray_image = cv2.imread('grayscale_image.jpg')

# Apply Simple Binary Thresholding (Thresh=125)
_, binThresh = cv2.threshold(gray_image, 125, 255, cv2.THRESH_BINARY)
# Display the edited image in a window
cv2_imshow(binThresh)
```

```python
# Apply Inverted Binary Thresholding (Thresh=125)
_, binInvThresh = cv2.threshold(gray_image, 125, 255, cv2.THRESH_BINARY_INV)
# Display the edited image in a window
cv2_imshow(binInvThresh)
```

```python
# Apply Truncated Thresholding (Thresh=125)
_, truncThresh = cv2.threshold(gray_image, 125, 255, cv2.THRESH_TRUNC)

# Display the edited image in a window
cv2_imshow(truncThresh)
```

```python
# Apply To Zero Thresholding (Thresh=125)
_, toZeroThresh = cv2.threshold(gray_image, 125, 255, cv2.THRESH_TOZERO)
# Display the edited image in a window
cv2_imshow(toZeroThresh)
```

```python
# Apply To Zero Inverted Thresholding (Thresh=125)
_, toZeroInvThresh = cv2.threshold(gray_image, 125, 255, cv2.THRESH_TOZERO_INV)
# Display the edited image in a window
cv2_imshow(toZeroInvThresh)
```

```python
# Count the number of pixels in range 0-255
gray_hist = cv2.calcHist([gray_image],[0],None,[256],[0,256])

# Show the Plotting Graph of the Grayscale Image
plt.plot(gray_hist)
plt.show()
```

```python
# Count the number of pixels in range 0-255
blue_hist = cv2.calcHist([image],[0],None,[256],[0,256])
green_hist = cv2.calcHist([image],[1],None,[256],[0,256])
red_hist = cv2.calcHist([image],[2],None,[256],[0,256])

# Show the Plotting Graph of the Colored Image
plt.plot(red_hist)
plt.plot(green_hist)
plt.plot(blue_hist)
plt.show()
```
